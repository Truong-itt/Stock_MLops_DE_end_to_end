```python
import asyncio
import yfinance as yf
from datetime import datetime

async def main():

    def message_handler(message):
        # In raw message nhбє­n Д‘Ж°б»Јc tб»« WebSocket
        print("рџ“© RAW MESSAGE:")
        print(message)

        # Nбєїu cГі timestamp thГ¬ in thГЄm dбєЎng readable
        if "time" in message and message["time"]:
            ts = int(message["time"])
            readable_time = datetime.fromtimestamp(ts / 1000).strftime('%Y-%m-%d %H:%M:%S')
            print(f"вЏ° Time: {readable_time}")

        print("-" * 60)

    # Kбєїt nб»‘i WebSocket Yahoo Finance
    ws = yf.AsyncWebSocket()

    # Danh sГЎch mГЈ cб»• phiбєїu
    stock_list = ["AAPL", "MSFT", "NVDA", "TSLA", "VCB", "FPT"]

    # Subscribe
    await ws.subscribe(stock_list)

    print("рџљЂ Subscribed, waiting for data...\n")

    # LбєЇng nghe dб»Ї liб»‡u
    await ws.listen(message_handler=message_handler)

if __name__ == "__main__":
    try:
        asyncio.run(main())
    except KeyboardInterrupt:
        print("\nрџ›‘ Stopped")
```

```python
from datetime import datetime
import pytz

def is_weekday(dt):
    return dt.weekday() < 5  # Thб»© 2вЂ“6

markets = {
    "рџ‡єрџ‡ё USA (NYSE/NASDAQ)": {
        "timezone": "America/New_York",
        "open": (9, 30),
        "close": (16, 0)
    },
    "рџ‡Єрџ‡є Europe (London)": {
        "timezone": "Europe/London",
        "open": (8, 0),
        "close": (16, 30)
    },
    "рџ‡Їрџ‡µ Japan (Tokyo)": {
        "timezone": "Asia/Tokyo",
        "open": (9, 0),
        "close": (15, 0)
    },
    "рџ‡Ёрџ‡і China (Shanghai)": {
        "timezone": "Asia/Shanghai",
        "open": (9, 30),
        "close": (15, 0)
    },
    "рџ‡»рџ‡і Vietnam (HOSE)": {
        "timezone": "Asia/Ho_Chi_Minh",
        "open": (9, 0),
        "close": (15, 0)
    }
}

def check_open_markets():
    print("рџЊЌ GLOBAL MARKET STATUS")
    print("=" * 40)

    for name, info in markets.items():
        tz = pytz.timezone(info["timezone"])
        now = datetime.now(tz)

        open_time = now.replace(hour=info["open"][0], minute=info["open"][1], second=0)
        close_time = now.replace(hour=info["close"][0], minute=info["close"][1], second=0)

        is_open = (
            is_weekday(now) and
            open_time <= now <= close_time
        )

        status = "рџџў ДђANG GIAO Dб»ЉCH" if is_open else "рџ”ґ ДђГ“NG Cб»¬A"
        print(f"{name}: {status} | Local time: {now.strftime('%H:%M:%S')}")

if __name__ == "__main__":
    check_open_markets()
```
